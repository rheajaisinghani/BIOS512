````r
options(repr.plot.width = 8, repr.plot.height = 3)

p = ggplot(penguins_long, aes(x = species, y = value, color = sex, position_jitter)) +

    geom_point(size = 0.9, alpha = 0.6, position = position_jitter(w = 0.3, h = 0.1)) +

    facet_wrap(~ name, nrow = 1, scales = "free") +

    stat_summary(fun = median, 
                 fun.min = median, 
                 fun.max = median, 
                 geom = "crossbar", 
                 width = 0.75, 
                 size = 0.2,
                 mapping = aes(group = species)) +

    theme(axis.text.x = element_text(hjust = 1, angle = 45))

p
````
